```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('D:/ExcelR/A14 decision trees/Company_Data.csv')
```

```python
dataset['Urban'].replace({'Yes':1,'No':0},inplace=True)
dataset['US'].replace({'Yes':1,'No':0},inplace=True)
```

```python
dataset.head()
```

```python
dataset['ShelveLoc'].replace({'Bad':1,'Medium':2,'Good':3},inplace=True)
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
bins = [-0.1, 5.4, 7.5, 9.3, 16.3, np.inf]
names = [1, 2, 3, 4, 5]
dataset['Sales_category'] = pd.cut(dataset['Sales'], bins, labels=names)
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
